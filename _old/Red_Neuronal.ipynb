```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import os
```

```python





# Cargar los datos en lotes y realizar transformaciones
def load_and_process_data(csv_file):
    # Cargar el lote de datos desde el archivo CSV
    data = pd.read_csv(csv_file)
    data.dropna(subset=['num_docks_available', 'Llocs'], inplace=True)
    data = data[data["num_docks_available"] >= 0]
    data = data[data["Llocs"] >= 0]
    data = data[data["num_docks_available"] <= 60]
    data = data[data["Llocs"]  <= 60]
    data = data[data["num_docks_available"]  <= data["Llocs"]]
    print(data["num_docks_available"].max())
    print(data["Llocs"].max())
    # Realizar las transformaciones necesarias en los datos
    # Por ejemplo, codificar variables categóricas, escalar valores numéricos, etc.

    # Dividir los datos en características (X) y etiquetas (Y)
    # X = data[['station_id','hour','dayofweek','month','Wind','Rain']]
    # X = data[['station_id','hour','dayofweek','month']]
    X = data[['station_id','hour','dayofweek']]

    Y = (data['num_docks_available']/data['Llocs']).fillna(0)

    # Devolver los datos procesados
    return X, Y
def load_and_process_data2(csv_file):
    # Cargar el lote de datos desde el archivo CSV
    data = pd.read_csv(csv_file)
    # Realizar las transformaciones necesarias en los datos
    # Por ejemplo, codificar variables categóricas, escalar valores numéricos, etc.

    # Dividir los datos en características (X) y etiquetas (Y)
    X = data[['station_id','hour','dayofweek']]
    # Devolver los datos procesados
    return X

fileList = os.listdir('/content/drive/MyDrive/prova/dades')
csv_files = [file for file in fileList if file[-3:] == "csv"]
batch_size = 1000

# Crear un modelo de red neuronal
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(3,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

# Compilar el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Escalador para el escalado por lotes
# scaler = MinMaxScaler()

# Iterar sobre los archivos CSV y entrenar el modelo por lotes
for csv_file in csv_files[:]:
    # Cargar los datos en el lote
    X_train, Y_train = load_and_process_data('/content/drive/MyDrive/prova/dades/'+csv_file)

    # Dividir los datos en conjuntos de entrenamiento y prueba
    # X_train, X_test, Y_train, Y_test = train_test_split(X_train, Y_train, test_size=0.2)

    # Escalar los valores numéricos en el rango [0, 1] utilizando el escalador por lotes
    # X_train_scaled = scaler.fit_transform(X_train)
    X_train_scaled =X_train
    # X_test_scaled = scaler.transform(X_test)
    # X_test_scaled = X_test
    # Entrenar el modelo en el lote actual
    model.fit(X_train_scaled, Y_train, epochs=4, batch_size=32)

    # Evaluar el rendimiento del modelo en el conjunto de prueba
    # loss = model.evaluate(X_test_scaled, Y_test)
    # print(f"Loss en el lote {csv_file}: {loss}")
    print(csv_file)

```
